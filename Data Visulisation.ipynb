```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(10,4),facecolor='Gray')
a=np.random.randint(1,20,30)

plt.plot(a,"ro--",lw=2,label="Random Plot")
plt.title("RANDOM PLOT",fontsize=20)
plt.xlabel("X-AXIS",fontsize=20)
plt.ylabel("Y-AXIS",fontsize=20)
plt.xticks(np.arange(0,30))
plt.yticks(np.arange(0,20))
plt.legend()
plt.savefig("randomplot.jpeg")
plt.show()
```

```python
plt.figure(figsize=(10,6),facecolor="gray")

angle=np.arange(0,361,15)
rad=np.deg2rad(angle)
sin=np.sin(rad)
cos=np.cos(rad)
tan=np.tan(rad)
plt.plot(angle,sin,"ro--",lw=2,label="SIN$\Theta$",alpha=0.9)
plt.plot(angle,cos,"go--",lw=2,label="COS$\Theta$",alpha=0.9)
plt.xlabel("ANGLE",fontsize=20)
plt.ylabel("RADIAN",fontsize=20)
plt.xticks(np.arange(0,361,30))
#plt.yticks(rad)
plt.title("SIN VS COS",fontsize=20)
plt.legend(fontsize=15)
plt.savefig("SINVSCOS.png")
#plt.plot(angle,tan)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,6),facecolor="Gray")
x=np.linspace(-5,5,500)
ax.plot(x,x**2,"r--",lw=3,label="$X^2$")
ax.plot(x,x**3,"g--",lw=3,label="$X^3$")
ax.set_facecolor("#123456")
ax.set_xlabel("X-Axis",fontsize=20)
ax.set_ylabel("Y-Axis",fontsize=20)
ax.set_xticks(np.arange(-10,10,2))
ax.set_yticks(np.arange(-120,120,20))
ax.legend(fontsize=20)
ax.set_title("X^2 VS X^3",fontsize=20)
ax.set_xlim([-20,20])
ax.set_ylim([-150,150])
plt.show()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(8,3))
#print(ax)
fig.set_facecolor("#abcdef")
ax[2].plot(x,x**4)
ax[1].plot(x,x**3)
ax[0].plot(x,x**2)
plt.show()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(8,3))
power = 2
fig.set_facecolor("#abcdef")
title =  [ "$X^2$","$X^3$","$X^4$"]
ax[0].set_ylabel("Y_axis")
for axes in ax : 
    axes.plot(x,x**power)
    axes.spines['right'].set_visible(False)
    axes.spines['top'].set_visible(False)
    axes.set_facecolor("#abcdef")
    axes.set_title(title[power-2])
    axes.set_xlabel("X_axis")
    power += 1
plt.show()
```

```python
fig,ax = plt.subplots(nrows=3,ncols=2,sharex=True,sharey=True)
x = np.linspace(-5,5,100)
power = 2
for axes in ax :  
        axes[0].plot(x,x**power)
        axes[1].plot(x,(x+5)**power)
        power += 1
        axes[0].spines['right'].set_visible(False)
        axes[1].spines['right'].set_visible(False)
        axes[0].spines['top'].set_visible(False)
        axes[1].spines['top'].set_visible(False)
        break
        
```
